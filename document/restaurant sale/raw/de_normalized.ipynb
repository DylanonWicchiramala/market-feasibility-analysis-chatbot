```python
import pandas as pd

cen_df = pd.read_csv("fulfilment_center_info.csv")
meal_df = pd.read_csv("meal_info.csv")
df = pd.read_csv("train.csv")
```

```python
# Merge meal_df into df on the meal_id column
merged_df = pd.merge(df, meal_df, on='meal_id')

# Display the resulting DataFrame
merged_df = pd.merge(merged_df, cen_df, on="center_id")

columns_to_remove = ["meal_id", "center_id", "checkout_price", "emailer_for_promotion", "homepage_featured", "center_type", "cuisine", "city_code", "region_code", "op_area"]
merged_df = merged_df.drop(columns=columns_to_remove)
```
